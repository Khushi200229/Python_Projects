# Data Analysis & Visualization

```python
#Find the cancellation rate.....
#Finding canceled percentage

canceled_perc = hotel_df['is_canceled'].value_counts(normalize=True) 
print(canceled_perc)
# normalize=True(relative frequencies (percentages) of each unique value)
# normalize=False(Default)( returns the raw counts of each unique value)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not Canceled','Canceled'],hotel_df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=hotel_df, palette='Blues')
legend_labels,_ = ax1. get_legend_handles_labels() # legend_labels,Legend_handles(_ for discard)
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('no. of reservations')
plt.show()
```

```python
#Cancelation percentage ratio of 'City Hotel'
city_hotel_df = hotel_df[hotel_df['hotel'] == 'City Hotel']
city_hotel_df['is_canceled'].value_counts(normalize=True)
```

```python
#Cancelation percentage ratio of 'Resort Hotel'
resort_hotel_df = hotel_df[hotel_df['hotel'] == 'Resort Hotel']
resort_hotel_df['is_canceled'].value_counts(normalize=True)
```

```python
#finding mean of 'adr' using groupby 'reservation_status_date' column
city_hotel_df = city_hotel_df.groupby('reservation_status_date')[['adr']].mean()
city_hotel_df
```

```python
#finding mean of 'adr' using groupby 'reservation_status_date' column
resort_hotel_df = resort_hotel_df.groupby('reservation_status_date')[['adr']].mean()
#resort_hotel_df.to_string()
#pd.set_option('display.max_rows',None)
#resort_hotel_df.to_markdown
resort_hotel_df
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City Hotel', fontsize=30)
plt.plot(resort_hotel_df.index,resort_hotel_df['adr'], label='Resort Hotel')
plt.plot(city_hotel_df.index,city_hotel_df['adr'], label='City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
hotel_df['month'] = hotel_df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month', hue='is_canceled', data = hotel_df, palette='bright')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month', size=20)
plt.xlabel('month')
plt.ylabel('number of reservation')
#plt.legend('not canceled','canceled')
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.title('ADR per month', fontsize=30)
#Seaborn in fact has six variations of matplotlib’s palette, called deep, muted, pastel, bright, dark, and colorblind.(rocket,vlag)
sns.barplot(x ='month',y='adr', data=hotel_df[hotel_df['is_canceled']==1].groupby('month')[['adr']].sum(), palette="bright")
plt.legend(fontsize=20)
plt.show()
```

```python
#Creating new dataframe with name 'canceled_data' which contains only canceled booking records
canceled_data = hotel_df[hotel_df['is_canceled']==1]

#Creating new dataframe with name 'not canceled_data' which contains only canceled booking records
not_canceled_data = hotel_df[hotel_df['is_canceled']==0]

top_10_country = canceled_data['country'].value_counts()[:10]
print(top_10_country)

plt.figure(figsize= (8,8))
plt.title('Top 10 countries with canceled reservation')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled= hotel_df[hotel_df['is_canceled']==0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize =(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='canceled')
plt.legend()
plt.show()

```
